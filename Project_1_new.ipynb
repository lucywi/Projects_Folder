```python
#Question3.13
#Find the standard deviation
def stdev(dict_values):
    mean = sum(dict_values)/len(dict_values)
    sq_distances = [(i-mean)**2 for i in dict_values]
    avg_sq_distances = sum(sq_distances)/len(dict_values)
    stand_dev = avg_sq_distances**.5
    return stand_dev
    
    
print('The rate standard deviation is: %.6r') % (stdev(rate_dict.values()))
print('The verbal standard deviation is: %.6r') % (stdev(verbal_dict.values()))
print('The math standard deviation is: %.6r') % (stdev(math_dict.values()))
```
