```python
import pandas as pd

#read in the raw star wars data set by the url

star_wars = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/star-wars-survey/StarWars.csv')

star_wars_head = star_wars.head(2)
print star_wars_head
print'\n'
print('The head funcion allows us to see the first few rows of the data set (the default \nnumber of rows is 5 but you can specify otherwise)')
print'\n'
print'\n'

star_wars_tail = star_wars.tail(2)
print star_wars_tail
print'\n'
print('The tail function allows us to see the last few rows of the data set, just \nlike the head function does for the first rows')
print'\n'
print'\n'

star_wars_summary = star_wars.describe()
print star_wars_summary
print'\n'
print('The describe function shows us some of the summary statistics (count, mean, std, etc...) \n for the dataset')


```
